```python
# Import the IMDB dataset from Keras
from tensorflow.keras.datasets import imdb

# Load the dataset
# num_words=None loads all words, not just the top N most frequent ones
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=None)
```

## Task 2: Text Preprocessing

### 1. Decode reviews

```python
# Get the word index (word -> integer mapping)
word_index = imdb.get_word_index()

# Create a reverse word index (integer -> word mapping)
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

# The indices are offset by 3 because 0, 1, and 2 are reserved indices for 'padding', 'start of sequence', and 'unknown'
def decode_review(text_integers):
    return ' '.join([reverse_word_index.get(i - 3, '?') for i in text_integers])

# Decode a sample training review
print("Original (encoded) training review:", train_data[0][:20])
decoded_review = decode_review(train_data[0])
print("Decoded training review (first 1000 chars):\n", decoded_review[:1000])
```

```python
import re
import nltk
from nltk.corpus import stopwords

# Download NLTK stopwords (if not already downloaded)
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def clean_text(text):
    # 1. Lowercase conversion
    text = text.lower()

    # 2. HTML tag removal
    text = re.sub(r'<.*?>', '', text)

    # 3. Punctuation removal
    text = re.sub(r'[^a-z\s]', '', text) # Keep only letters and spaces

    # 4. Stopword removal
    words = text.split()
    words = [w for w in words if w not in stop_words]
    text = ' '.join(words)

    return text

# Apply cleaning to decoded reviews
# For demonstration, let's clean the first decoded review
cleaned_sample_review = clean_text(decode_review(train_data[0])) # Corrected 'decoding_review' to 'decode_review'
print("Cleaned sample review (first 500 chars):\n", cleaned_sample_review[:500])

# Apply cleaning to all training and testing reviews
# This might take some time
print("Cleaning all training reviews...")
cleaned_train_reviews = [clean_text(decode_review(review)) for review in train_data]
print("Cleaning all test reviews...")
cleaned_test_reviews = [clean_text(decode_review(review)) for review in test_data]

print("Finished cleaning reviews.")
```

### 3. Tokenization and Sequence Generation

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Define vocabulary size
VOCAB_SIZE = 10000

# Initialize tokenizer
# num_words limits the vocabulary size to the most frequent VOCAB_SIZE words
tokenizer = Tokenizer(num_words=VOCAB_SIZE, oov_token='<unk>')

# Fit tokenizer on training data (cleaned reviews)
tokenizer.fit_on_texts(cleaned_train_reviews)

# Convert text reviews to sequences of integers
print("Converting training reviews to sequences...")
train_sequences = tokenizer.texts_to_sequences(cleaned_train_reviews)
print("Converting test reviews to sequences...")
test_sequences = tokenizer.texts_to_sequences(cleaned_test_reviews)

print("First training review (original decoded):")
print(cleaned_train_reviews[0][:100]) # Print first 100 characters
print("First training review (tokenized sequence):")
print(train_sequences[0][:20]) # Print first 20 tokens
```

### 4. Padding

```python
# Define maximum sequence length (based on earlier analysis, average was ~235)
MAX_LEN = 250

# Pad sequences to ensure uniform length
# 'post' padding means padding zeros at the end
# 'truncating=post' means remove values from the end of sequences longer than maxlen
print(f"Padding training sequences to max length {MAX_LEN}...")
padded_train_sequences = pad_sequences(train_sequences, maxlen=MAX_LEN, padding='post', truncating='post')
print(f"Padding test sequences to max length {MAX_LEN}...")
padded_test_sequences = pad_sequences(test_sequences, maxlen=MAX_LEN, padding='post', truncating='post')

print("Shape of padded training data:", padded_train_sequences.shape)
print("Shape of padded test data:", padded_test_sequences.shape)

print("First padded training sequence (first 20 elements):")
print(padded_train_sequences[0][:20])
```

## Task 3: Build SimpleRNN Model

### 1. Build the SimpleRNN Model Architecture

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

# Define model parameters
embedding_dim = 32 # Dimension of the word embeddings
rnn_units = 32     # Number of units in the SimpleRNN layer

# Build the SimpleRNN model
simple_rnn_model = Sequential([
    Embedding(input_dim=VOCAB_SIZE, output_dim=embedding_dim, input_length=MAX_LEN),
    SimpleRNN(units=rnn_units),
    Dense(units=1, activation='sigmoid') # Output layer for binary classification
])

# Display the model summary
simple_rnn_model.summary()
```

### 2. Compile the Model

```python
simple_rnn_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

### 3. Train the Model

```python
epochs = 10
batch_size = 32

print("Training SimpleRNN model...")
history_simple_rnn = simple_rnn_model.fit(
    padded_train_sequences,
    train_labels,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.2 # Use 20% of training data for validation
)
```

### 4. Evaluate the Model

```python
print("Evaluating SimpleRNN model on test data...")
loss_simple_rnn, accuracy_simple_rnn = simple_rnn_model.evaluate(padded_test_sequences, test_labels)

print(f"SimpleRNN Test Loss: {loss_simple_rnn:.4f}")
print(f"SimpleRNN Test Accuracy: {accuracy_simple_rnn:.4f}")
```

## Task 4: Build LSTM Model

### 1. Build the LSTM Model Architecture

```python
from tensorflow.keras.layers import LSTM

# Build the LSTM model
lstm_model = Sequential([
    Embedding(input_dim=VOCAB_SIZE, output_dim=embedding_dim, input_length=MAX_LEN),
    LSTM(units=rnn_units), # Using LSTM layer instead of SimpleRNN
    Dense(units=1, activation='sigmoid')
])

# Display the model summary
lstm_model.summary()
```

### 2. Compile the Model

```python
lstm_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

### 3. Train the Model

```python
print("Training LSTM model...")
history_lstm = lstm_model.fit(
    padded_train_sequences,
    train_labels,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.2
)
```

### 4. Evaluate the Model

```python
print("Evaluating LSTM model on test data...")
loss_lstm, accuracy_lstm = lstm_model.evaluate(padded_test_sequences, test_labels)

print(f"LSTM Test Loss: {loss_lstm:.4f}")
print(f"LSTM Test Accuracy: {accuracy_lstm:.4f}")
```

## Task 5: Build GRU Model

### 1. Build the GRU Model Architecture

```python
from tensorflow.keras.layers import GRU

# Build the GRU model
gru_model = Sequential([
    Embedding(input_dim=VOCAB_SIZE, output_dim=embedding_dim, input_length=MAX_LEN),
    GRU(units=64), # Increased units from 32 to 64
    Dense(units=1, activation='sigmoid')
])

# Display the model summary
gru_model.summary()
```

### 2. Compile the Model

```python
gru_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

### 3. Train the Model

```python
print("Training GRU model...")
history_gru = gru_model.fit(
    padded_train_sequences,
    train_labels,
    epochs=5,
    batch_size=batch_size,
    validation_split=0.2
)
```

### 4. Evaluate the Model

```python
print("Evaluating GRU model on test data...")
loss_gru, accuracy_gru = gru_model.evaluate(padded_test_sequences, test_labels)

print(f"GRU Test Loss: {loss_gru:.4f}")
print(f"GRU Test Accuracy: {accuracy_gru:.4f}")
```

### 1. Calculate Additional Metrics (Precision, Recall, F1 Score)

```python
from sklearn.metrics import precision_score, recall_score, f1_score
import time

def evaluate_model_metrics(model, test_sequences, test_labels, threshold=0.5):
    predictions = model.predict(test_sequences)
    binary_predictions = (predictions > threshold).astype(int)

    precision = precision_score(test_labels, binary_predictions)
    recall = recall_score(test_labels, binary_predictions)
    f1 = f1_score(test_labels, binary_predictions)

    return precision, recall, f1

print("Calculating additional metrics for SimpleRNN...")
precision_simple_rnn, recall_simple_rnn, f1_simple_rnn = evaluate_model_metrics(simple_rnn_model, padded_test_sequences, test_labels)

print("Calculating additional metrics for LSTM...")
precision_lstm, recall_lstm, f1_lstm = evaluate_model_metrics(lstm_model, padded_test_sequences, test_labels)

print("Calculating additional metrics for GRU...")
precision_gru, recall_gru, f1_gru = evaluate_model_metrics(gru_model, padded_test_sequences, test_labels)
```

## Task 8: Real-Time Testing with Custom Reviews

```python
def predict_sentiment(review_text, model, tokenizer, max_len, clean_text_func):
    # 1. Clean the review text
    cleaned_review = clean_text_func(review_text)

    # 2. Tokenize the cleaned review
    sequence = tokenizer.texts_to_sequences([cleaned_review])

    # 3. Pad the sequence
    padded_sequence = pad_sequences(sequence, maxlen=max_len, padding='post', truncating='post')

    # 4. Make prediction
    prediction = model.predict(padded_sequence)[0][0] # Get the probability for the positive class

    # 5. Determine sentiment and confidence
    if prediction >= 0.5:
        sentiment = "Positive"
        confidence = prediction
    else:
        sentiment = "Negative"
        confidence = 1 - prediction # Confidence for negative sentiment

    return sentiment, confidence

# Example reviews provided by the user
custom_reviews = [
    "This movie was absolutely fantastic and I enjoyed every minute.",
    "The movie was boring and a complete waste of time.",
    "The acting was good but the story was average."
]

models = {
    "SimpleRNN": simple_rnn_model,
    "LSTM": lstm_model,
    "GRU": gru_model
}

# Test each model with custom reviews
for model_name, model_obj in models.items():
    print(f"\n--- {model_name} Model Predictions ---")
    for i, review in enumerate(custom_reviews):
        sentiment, confidence = predict_sentiment(review, model_obj, tokenizer, MAX_LEN, clean_text)
        print(f"Review {i+1}: '{review}'")
        print(f"  Predicted Sentiment: {sentiment}")
        print(f"  Confidence Score: {confidence:.4f}")
```
